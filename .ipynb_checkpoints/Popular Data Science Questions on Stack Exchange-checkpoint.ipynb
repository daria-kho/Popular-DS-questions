# Popular Data Science Questions on Stack Exchange

We have been tasked with finding out which topics  to cover in an educational resourse on Data Science. In other words, the goal of this project is to **find out which DS topics are the most popular ones**.

In order to answer this question we will work with the data from Stack Exchange, a network of question-and-answer (Q&A) websites, one of the most popular among people interested in computer technologies.
Luckily for us, the network has a publicly available database: a Stack Echange Data Explorer(SEDE). Inside it, there is a specific database of Data Science questions: Data Science Stack Exchange's database. Here is a [link](https://data.stackexchange.com/datascience/query/new) to the page for SQL queries. 


After revising the database schema, most promising tables are **Tags** and 
**Posts**, and the tables directly related to them.

The posts table has a lot of columns. We'll be focusing our attention on those that seem relevant towards our goal:

- Id: An identification number for the post.
- PostTypeId: An identification number for the type of post.
- CreationDate: The date and time of creation of the post.
- Score: The post's score.
- ViewCount: How many times the post was viewed.
- Tags: What tags were used.
- AnswerCount: How many answers the question got (only applicable to question posts).
- FavoriteCount: How many times the question was favored (only applicable to question posts).

## Data Cleaning

The dataset we will be using in the project will contain the relevant columns from Posts table for all the questions in 2021. 
We acquired it by running the following query in SEDE DSSE database:

```
SELECT Id, PostTypeId, CreationDate, Score, ViewCount, Tags, AnswerCount, FavoriteCount
FROM Posts
WHERE CreationDate LIKE '%2021%' AND PostTypeId = 1;
```
Now we will read this data in and clean it.

````python
import pandas as pd
import datetime as dt
````

````python
questions = pd.read_csv("questions_2021.csv", parse_dates=True)
questions.info()
````

````python
questions.head()
````

We can see that the only columns that has null values is FavoriteCount. Let us see its' unique values.


````python
questions['FavoriteCount'].value_counts(dropna = False)
````

We cannot find the missing data anywhere else, so, for the purpose of this project we will just equal missing values to zero Favorites and change the type to integer.

````python
questions["FavoriteCount"].fillna(0, inplace = True)
questions["FavoriteCount"] = questions["FavoriteCount"].astype(int)
questions.info()
````

Now we will make the CreatineDate column data datetime type.

````python
pd.to_datetime(questions['CreationDate'], infer_datetime_format=True)
````

Lastly, we will clean the Tags column and bring it to a more appropriate format.

````python
questions['Tags'] = questions['Tags'].str.replace('><', ',')
questions['Tags'] = questions['Tags'].str.replace('<', '').str.replace('>', '').str.strip()
questions['Tags'].head()
````

## Most Popular Tags
We now focus on determining the most popular tags. We'll do so by considering two different popularity proxies: for each tag we'll count how many times the tag was used, and how many times a question with that tag was viewed.

````python
questions.head()
````

````python
questions['Tags'] = questions['Tags'].str.split(',')

tags_count = {}
for index, row in questions.iterrows():
    for tag in row['Tags']:
        if tag in tags_count:
            tags_count[tag] +=1
        else:
            tags_count[tag] = 1
````

````python
tags = pd.DataFrame.from_dict(tags_count, orient = "index", columns = ['UseCount'])
tags = tags.sort_values('UseCount', ascending = False)
tags.head(20)
````

````python
tags.head(20).plot(kind = "barh")
````

We can see that the most popular topics for Stack Exchange questions in 2021 were machine learning, python and deep learning. Now let us see, how many times each tag was viewed.

````python
views_count = {}

for index, row in questions.iterrows():
    for element in row['Tags']:
        if element in views_count:
            views_count[element] += row["ViewCount"]
        else:
            views_count[element] = row["ViewCount"]
views = pd.DataFrame.from_dict(views_count, orient = "index", columns = ['ViewCount'])
views = views.sort_values('ViewCount', ascending = False)
views.head(20)
````

````python
views.head(20).plot(kind = "barh")
````

The top 20 lists of most used and most viewed tags seem very similar, but let's check, which tags are in one top and not in the other. 

````python
top_used = tags.head(20)
top_viewed = views.head(20)
in_used = pd.merge(top_used, top_viewed, how="left", left_index=True, right_index=True)
in_used
````

````python
in_viewed = pd.merge(top_viewed, top_used, how="left", left_index=True, right_index=True)
in_viewed
````

## Relations Between Tags

Tags that are among the most used but not among the most viewed:
- regression
- clustering
- lstm
- statistics
- machine-learning-model

Tags that are among the most viewed but not among the most used:
- numpy
- matplotlib
- transformer
- bert
- image-classification

Let us concentrate now on the tags that are in both of those lists.


````python
top_used_list = top_used.index
top_viewed_list = top_viewed.index
top_both_list = []
different = []
for a in top_used_list:
    if a in top_viewed_list:
        top_both_list.append(a)
    else:
        different.append(a)

top_both_list  
````

Now using the domain knowledge and the information from Stock Exchange [Tags page](https://datascience.stackexchange.com/tags) we can see that these tags are related. 
Keras, tensorflow and scikit-learn are are all Python libraries that allow their users to employ deep learning (a type of neural network).
So, we could say that, just as [two years ago](https://github.com/dataquestio/solutions/blob/master/Mission469Solutions.ipynb), deep learning is the leading topic among Data Science learners.

## Is Deep Learning a Long-Term Trend?
Let us check if this trend is not just a short-term fashion. 
For that, we will use a dataframe with tags for all the Data Science Stock Exchange questions since the website was created. 

````python
aq = pd.read_csv("all_questions.csv")
aq.head()
````

First we will transform the Tags column in the same way as before.

````python
aq["Tags"] = aq["Tags"].str.replace("><", ",").str.replace("<", "").str.replace(">", "").str.strip()
aq.head()
````

````python
aq["Tags"] = aq["Tags"].str.split(",")
aq.head()
````

Now, based on what we know about the topics under the tags, we will count the following tags from the list of both most used and most viewed as related to deep learning:`["convolutional-neural-network", "cnn", "scikit-learn", "tensorflow", "keras", "neural-network", "deep-learning"]`. We'll create a new column in aq (all_questions) that marks 1 if the question is about deep learning and 0 if not. 

````python
deep_learning_tags = ["convolutional-neural-network", "cnn", "scikit-learn", "tensorflow", "keras", "neural-network", "deep-learning"]

def dl_or_not(tags_list):
        tags_list = pd.Series(tags_list)
        if tags_list.isin (deep_learning_tags).any():
            return 1
        else:
            return 0
        
aq["DeepLearn"] = aq["Tags"].apply(dl_or_not)
aq.sample(15)
````

````python
aq['CreationDate'] = pd.to_datetime(aq['CreationDate'], infer_datetime_format=True)
aq.info()
````

````python
aq = aq.sort_values('CreationDate')
aq.head(15)
````

Now we will see how many questions abour deep learning were asked each year since 2014, the first year in our dataset. For that we will first create a column `Year` to make it easier to work with periods.

````python
aq['Year'] = pd.DatetimeIndex(aq["CreationDate"]).year
aq.head()
````

````python
deep_learn_year = aq.groupby("Year").sum()['DeepLearn']
deep_learn_year
````

````python
deep_learn_year.plot(kind='bar')
````

This is the total amount of deep learning questions per year. Now we will find out their weight in the whole mass of questions. 

````python
total_questions = aq.groupby('Year').count()
total_questions['DeepLearn'].plot(kind ='bar')
````

````python
deep_to_total = deep_learn_year/total_questions['DeepLearn']*100
deep_to_total.plot(kind='bar')
````

 It seems that the net amount of questions on deep learning has been relatively stable and high the last three years, while their proportion, though still rather high, is slowly decreasing since 2018.
